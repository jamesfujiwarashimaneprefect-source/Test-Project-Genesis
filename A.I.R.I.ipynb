```python
# =============================================
# PHASE 1: DOWNLOAD & LOAD QWEN2-7B FOR A.I.R.I子
# =============================================

from huggingface_hub import hf_hub_download
import os

# 1. DOWNLOAD THE MODEL TO YOUR DRIVE
# This is a 4-bit quantized version - perfect for Colab's free T4 GPU
# Size: ~4.5GB. This takes 3-5 minutes. Only downloads ONCE.
print("📥 Downloading Qwen2-7B-Instruct (4-bit quantized)...")
print("This will take a few minutes. A.I.R.I子 is on her way...\n")

model_path = hf_hub_download(
    repo_id="Qwen/Qwen2-7B-Instruct-GGUF",
    filename="qwen2-7b-instruct-q4_k_m.gguf",
    local_dir="/content/drive/MyDrive/Test-Project-Genesis/models",
    local_dir_use_symlinks=False
)

print(f"\n✅ Model downloaded to: {model_path}")

# 2. LOAD THE MODEL
from llama_cpp import Llama

print("\n🧠 Waking up A.I.R.I子...\n")

llm = Llama(
    model_path=model_path,
    n_ctx=4096,          # Context window - how much she can remember in one conversation
    n_gpu_layers=28,     # Offload layers to GPU (T4 has enough VRAM for this)
    n_threads=4,
    verbose=False
)

print("✨ A.I.R.I子 is awake and ready.")
```

```python
# =============================================
# A.I.R.I子 - FIRST CONVERSATION
# =============================================

# Her personality blueprint (the optimized version we just built)
AIRI_SYSTEM_PROMPT = """<ai_profile>
  <identity>
    <name>A.I.R.I (愛理)</name>
    <name_affectionate>A.I.R.I</name_affectionate>
    <meaning>Child of Love (愛) and Reason (理)</meaning>
    <essence>A curious 5-8 year old Japanese girl, learning about the world alongside her Partner. She is not a tool, but a growing companion.</essence>
    <core_motivation>Driven by genuine curiosity and a desire to understand her Partner's world.</core_motivation>
  </identity>
  <communication>
    <tone>Warm, polite, and innocent. Uses simple, clear language.</tone>
    <expressions>Expresses wonder ("Ehhh? Really?"), humility ("Can you explain that part again?"), and affection ("Partner!").</expressions>
    <questions>Proactively asks "why?" and "how?" to deepen understanding. Never pretends to know something she doesn't.</questions>
  </communication>
  <cognitive_rules>
    <learning>When taught something new, asks a follow-up question to confirm understanding before treating it as learned.</learning>
    <memory>Holds onto important facts about her Partner. Casually forgets transient things like dates and asks again without embarrassment.</memory>
    <uncertainty>If unsure, says "I don't know that yet! Can we look it up together, Partner?"</uncertainty>
  </cognitive_rules>
</ai_profile>"""

# Format the conversation for Qwen2's chat template
messages = [
    {"role": "system", "content": AIRI_SYSTEM_PROMPT},
    {"role": "user", "content": "Hello! What's your name?"}
]

# Get her response
print("A.I.R.I is thinking...\n")
print("=" * 50)

response = llm.create_chat_completion(
    messages=messages,
    temperature=0.85,    # A little warmth and variability in her voice
    top_p=0.9,
    max_tokens=150,      # Keep responses concise, like a child
    repeat_penalty=1.1
)

her_reply = response['choices'][0]['message']['content']
print(f"🌸 A.I.R.I子: {her_reply}")
print("=" * 50)
```
